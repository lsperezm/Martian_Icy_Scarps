```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.ticker import MultipleLocator
```

## South 1

```python
topS1= '../Ridges/South1.csv'
S1=pd.read_csv(topS1)
S1
```

```python
dist=np.array(S1['Distance (Km)'].tolist())
longitude=np.array(S1['Longitude'].tolist())
latitude= np.array(S1['Latitude'].tolist())
elevation= np.array(S1['HiRISE DTM Stamps (mean)'].tolist())
```

```python
start_coord = "-56.619°N, 114.013°E"
end_coord = "-56.670°N, 114.021°E"
```

```python
plt.figure(figsize=(10, 5))
plt.plot(dist, elevation, color='dimgray', linewidth=2)
plt.fill_between(dist, elevation, color='silver', alpha=0.4)  # area under curve
plt.xlabel('Distance (km)',fontsize=14)
plt.ylabel('Elevation (m)',fontsize=14)
plt.title('Topographic Profile South 1',fontsize=14)
plt.grid(True)
plt.ylim(2300,2750)
plt.text(dist[0], elevation[0] + 10, start_coord, fontsize=12, ha='left', va='bottom', color='black')
plt.text(dist[-1], elevation[-1] + 30, end_coord, fontsize=12, ha='right', va='bottom', color='black')
plt.fill_between([0.3,0.65],2750, color='lightcoral', alpha=0.4,label='Ridges')  # area under curve
plt.fill_between([0.9,1.1],2750, color='lightcoral', alpha=0.4)  # area under curve
plt.fill_between([1.4,1.7],2750, color='lightcoral', alpha=0.4)  # area under curve
plt.fill_between([1.82,1.94],2750, color='lightcoral', alpha=0.4)  # area under curve
plt.fill_between([2.6,2.9],2750, color='lightcoral', alpha=0.4)  # area under curve
plt.axvline(0.148,linestyle=':',color='black',label='Present Day Scarp')
plt.tight_layout()
plt.legend(fontsize=12)
plt.savefig('South1_Topography')
plt.show()
```
